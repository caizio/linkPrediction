```python
# 通过计算面积的方法计算auroc,与抽样的方法计算，结果会有偏差
import matplotlib.pyplot as plt

list=[1,1,1,0,0,0,0,0,0]
tf_list=[0,1,1,0,0,0,0,0,0]
#给定一个分类器打分的列表，给定真实的正反例，4个正例(1,2,7,9)，6个反例(3,4,5,6,8,10)
#排序
#thresh=[1]
#thresh1=thresh+list
#print(thresh1)
m_true=0
m_flase=0
for i in range(len(tf_list)):
    if tf_list[i] == 1:  # 真正例
        m_true=m_true+1
    else:
        m_flase=m_flase+1
print(m_flase,m_true)        #6,4
x=[]
y=[]
xi=0
yi=0
for i in range(len(list)):
    if tf_list[i]==0:#假正例
        xi=xi+1/m_flase
        yi=yi
        x.append(xi)
        y.append(yi)
    else:#真正例
        xi=xi
        yi=yi+1/m_true
        x.append(xi)
        y.append(yi)
print(x,y)

plt.plot(x, y)

plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC_picture')
plt.show()###显示
AUC=0
for i in range(len(x)-1):
    AUC=AUC+(x[i+1]-x[i])*(y[i]+y[i+1])
auc=0.5*AUC
print(auc)
```
