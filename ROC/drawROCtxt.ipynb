```python
# ROCPR
# temp

import matplotlib.pyplot as plt
import readROCtxt as rd

dir = r"D:\work\\\SMS\\1 \2023_05_13 - .txt"
rROCt = rd.ReadROCTxt()
datas,datasName = rROCt.read(dir)
# 
# for i in range(6):
#     datas[i][6].name = "SMS_DCN"
#     datas[i][7].name = "DSMS_DCN"
#     datas[i][8].name = "maxSMS_DCN"
n = len(datas)
if n != 6:
    print("warning:")

# width = 20
# height = 12
width = 12
height = 18
fig = plt.figure(figsize=(width,height))
ax1= fig.add_subplot(3,2,1)
ax2= fig.add_subplot(3,2,2)
ax3= fig.add_subplot(3,2,3)
ax4= fig.add_subplot(3,2,4)
ax5= fig.add_subplot(3,2,5)
ax6= fig.add_subplot(3,2,6)
# axs = [ax1,ax2,ax3,ax4,ax5,ax6]
axs = [ax5,ax2,ax4,ax1,ax3,ax6]
# colors = ['lightcoral', 'coral', 'orange', 'gold', 'green', 'gray','blue','red','red']
colors = ['lightcoral', 'coral', 'orange', 'gold', 'green', 'gray','blue','darkmagenta','red']
for i in range(6):
    axs[i].set_xlabel("Recall")
    axs[i].set_ylabel("Precision")
    axs[i].set_title(datasName[i])
    axs[i].set_xlim(0.0,0.1)
    # axs[i].set_xlim(0,500)
    axs[i].set_ylim(0.0,1.0)
    j = 0
    for data in datas[i]:
        axs[i].plot(data.x, data.y,label = data.name,color=colors[j])
        j += 1
    axs[i].legend(loc = 1) 
plt.savefig("../../temp/6.pdf")
plt.show()
```
